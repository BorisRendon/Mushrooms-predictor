#### Checking null values

```python
samples = mushrooms.shape[0]
keep_cols = []
drop_cols = []

for col in mushrooms.columns:
    nas = mushrooms[col].isna().sum()
    percent = round(100*nas/samples,2)
    if nas > 0:
        print('{0:<20} {1:<20} {2}%'.format(col,nas,percent))
    if percent < 30:
        keep_cols.append(col)
    else:
        drop_cols.append(col)
```

En este dataset no tenemos columnas con valores nulos, lo cual nos favorece al no tener que hacer ningun tipo de imputación o tener que botar algún valor y enfocarnos a trabajarlo con los modelos.

### Split data

En este punto lo que tenemos que hacer es separar nuestra variable objetivo, que es **class** de las demás.

```python
mushrooms_class = DataFrame(mushrooms['class'])
```

```python
mushrooms_class.head()
```

```python
y=pd.Series(np.where(mushrooms_class['class'].values == 'p', 1, 0),
          mushrooms_class.index)
```

```python
y.head(5)
```

Como podemos observar toda la data con la que vamos a trabajar es data de tipo categórica, por lo cual vamos a tener que proceder a trabajar con OneHotEncoding

## One hot encoding the features

```python
encoder = OneHotEncoder()

encoded_mushrooms = mushrooms.drop('class',axis=1)

encoded_mushrooms1=DataFrame(encoder.fit_transform(encoded_mushrooms).toarray())
```

```python
encoded_mushrooms1
```

```python
print(y.shape,encoded_mushrooms.shape)
```

#### Training and test dataset

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(encoded_mushrooms1,y, test_size=0.35,random_state=42,shuffle=True)
```

Usando la librería de sklearn dividimos la data en entrenamiento y testing.

Para este caso elegí un test size del 30% del dataset y traning size del 70% con el hiperparámetro de *shuffle = True* , esto nos ayuda para poder evitar un posible overfitting del modelo.

```python
print(X_train.shape,y_train.shape)
```

```python
X_train.head()
```

### Regresión logística

La regresión logística estima la probabilidad que un dato pertenezca a una clase particular (binaria).

* Si la probabilidad es >0.5 entonces el output es 1.
* Si la probabilidad es <0.5 entonces el output es 0.

Por este motivo se acopla perfectamente al problema porque queremos ver si un hongo es venenoso(0) o comestible(1).



```python
from sklearn.linear_model import LogisticRegression
```

```python
logistic_R = LogisticRegression(solver='sag', random_state=42,C=5)
```

```python
logistic_R=logistic_R.fit(X_train,y_train)
```

```python
y_logistic_pred = logistic_R.predict(X_test)
```

```python
pd.DataFrame({'Actual': y_test,'Predictions': y_logistic_pred})
```

```python
log_reg = classification_report(y_test, y_logistic_pred, output_dict = True)

metrics_log_reg = pd.DataFrame(log_reg).transpose()

metrics_log_reg
```

 Entonces mis predicciones son 100% correctas cuando pronostico un caso positico. (1 en precisión) y logro detectar todos los casos positivos (1 recall).

### Redes neuronales

```python
import tensorflow as tf
from keras.models import Sequential
from keras.layers import Dense
```

##### Explicando funciones de activación

* Relu: esta función de acivación le asigna el valor de 0 si el resultado de z es negativo y le asigna el valor de z cuando z>0. Esta funcion de activación es de las más comunes actualmente.
![ReLu](imgs/relu.JPG)

*  Sigmoide: Esa función de activación se usa cuando hay que predecir la probabilidad de un output, es decir 0 o 1, por eso es adecuada para este modelo.
![Sigmoid](imgs/sigmoid.JPG)


##### Armando nuestra red neuronal.

Por ser un dataset sencillo decidí dejar solamente 3 capas.
* La capa inicial tiene 117 neuronas porque es la cantidad de features que tenemos disponibles y tiene la función de activación de *relu*.

* La segunda capa o capa oculta, tiene 3 neuronas y su función de activación también es de *relu*

* La tercera capa tiene solo una neurona porque recordemos que queremos predecir si un hongo es venenoso(1) o si es comestible (0), entonces es un output binario el que esperamos. Su función de activación es de sigmoide ya que como expliqué anteriormente, esperamos obtener un 1 o 0.


```python
model = Sequential()
model.add(Dense(117, input_dim=117, activation='relu'))
model.add(Dense(3, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

##### Tuneando nuestra red neuronal
Para el resultado final jugué con 2 parámetros los cuales son la función de *loss* y el optimizador.

Para el optimizador utilicé la función de **Adam**. 
* Adam a diferencia del stochastic gradient descent no deja el mismo learning rate a lo largo de todos los pesos, en cambio Adam, si va a cambiando el learning rate a lo largo del entrenamiento.

Para la función de loss utilicé la función de **binary cross entropy**.
* Esta función se acopla perfectamente a nuestra red porque nosotros queremos resultados que están entre 0 y 1. Esta funcion calcula un resultado que resume el promedio de la diferencia entre lo actual y lo predicho.


**Métrica para evaluar el modelo** 


La métrica para elegir es la de Recall ya que esta métrica es correcta cuando un falso negativo tiene un costo muy alto, y como en este proyecto estoy tratando de decir si un hongo es venenoso o no, sería peligroso clasificarlo como no venenoso cuando en realidad si lo es.

![Recall](imgs/recall.JPG)

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['Recall'])
```

El número de epochs es la cantidad de veces que el algoritmo va a correr y el número de batches es el numero que define las observaciones a trabajar antes de pasar al siguiente nodo.

```python
model.fit(X_train, y_train, epochs=50, batch_size=25)
```
